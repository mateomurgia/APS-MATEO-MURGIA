# Introducción

El análisis de circuitos en el dominio de la frecuencia permite estudiar cómo un sistema eléctrico responde ante señales senoidales de diferentes frecuencias. En lugar de observar el comportamiento instantáneo de tensiones y corrientes en el tiempo, se utiliza el dominio complejo de **Laplace** ($s$), donde los componentes pasivos —resistencias, inductancias y capacitancias— se representan mediante **impedancias y admitancias** dependientes de $s$. Esto facilita la obtención de la **función de transferencia**, herramienta fundamental para caracterizar el tipo de filtro y su respuesta en frecuencia.

En el dominio $s = \sigma + j\omega$, los elementos pasivos se modelan de la siguiente manera:

$$
Z_R = R, \quad Z_L = sL, \quad Z_C = \frac{1}{sC}
$$

y sus admitancias, inversas de la impedancia, son:

$$
Y_R = \frac{1}{R}, \quad Y_L = \frac{1}{sL}, \quad Y_C = sC
$$

A partir de estas relaciones se definen las **redes RLC** que, según su configuración (serie o paralelo), pueden comportarse como **filtros pasa bajos, pasa altos, pasa banda o rechaza banda**. Estos filtros modifican la amplitud y la fase de una señal de entrada según su frecuencia, determinando así su **respuesta en frecuencia**.

La función de transferencia general de un sistema lineal e invariante en el tiempo se expresa como:

$$
T(s) = \frac{V_{\text{salida}}(s)}{V_{\text{entrada}}(s)}
$$

Para los **filtros de segundo orden** (biquad o “sección biquadrática”), esta función toma la forma:

$$
T(s) = k \cdot \frac{s^2 + \frac{\omega_z}{Q_z}s + \omega_z^2}{s^2 + \frac{\omega_p}{Q_p}s + \omega_p^2}
$$

donde:

- $\omega_p$ y $\omega_z$ son las **frecuencias naturales** asociadas a los polos y ceros del sistema,  
- $Q_p$ y $Q_z$ son los **factores de calidad**, que indican el grado de selectividad o amortiguamiento,  
- y $k$ es la **ganancia estática**.

El estudio de estos parámetros permite predecir el comportamiento de la red ante distintas frecuencias, localizando los **polos y ceros en el plano $s$** para determinar su estabilidad y tipo de respuesta (resonante, amortiguada o plana).

y sus admitancias respectivas son:

$$
Y_R = \frac{1}{R}, \quad Y_L = \frac{1}{sL}, \quad Y_C = sC

A partir de estas relaciones se modela el circuito propuesto, compuesto por dos ramas paralelas asociadas a los coeficientes \( a, b, d \).  
La salida \( V_2 \) se toma sobre la segunda rama, mientras que la entrada \( V_1 \) se aplica al conjunto completo.

---

### Planteo de la función de transferencia

$$
T(s) = \frac{V_2}{V_1} = \frac{Z_2}{Z_1 + Z_2} = \frac{Y_{1,\text{total}}}{Y_{1,\text{total}} + Y_{2,\text{total}}}
$$

---

### Cálculo de admitancias parciales

$$
Y_{1,\text{total}} = \frac{s^2 RLCa + sLd + Rb}{sLR}
$$

$$
Y_{2,\text{total}} = \frac{s^2 RLC(1 - a) + sL(1 - d) + R(1 - b)}{sLR}
$$

---

### Admitancia total y \(T(s)\)

$$
Y_{1,\text{total}} + Y_{2,\text{total}} = \frac{s^2 RLC + sL + R}{sLR}
$$

$$
T(s) = \frac{s^2 RLCa + sLd + Rb}{s^2 RLC + sL + R}
$$

---

### Normalización

$$
T(s) = \frac{s^2 a + s\frac{d}{RC} + \frac{b}{LC}}{s^2 + s\frac{1}{RC} + \frac{1}{LC}}

## Ejercicio 2 — Parametrización de polinomios de segundo orden
La forma estándar de una sección biquadrática (SOS) es:

$$
T(s) = k \cdot \frac{s^2 + s\frac{\omega_z}{Q_z} + \omega_z^2}{s^2 + s\frac{\omega_p}{Q_p} + \omega_p^2}
$$

Comparando coeficiente a coeficiente se obtiene:

$$
\frac{d}{RLCa} = \frac{\omega_z}{Q_z}, \quad
\frac{b}{LCa} = \omega_z^2, \quad
\frac{1}{RC} = \frac{\omega_p}{Q_p}, \quad
\frac{1}{LC} = \omega_p^2
$$

---

### Frecuencias naturales

$$
\omega_z = \sqrt{\frac{b}{LCa}}, \quad
\omega_p = \sqrt{\frac{1}{LC}}
$$

---

### Factores de calidad

$$
Q_z = \frac{RCa}{d} \sqrt{\frac{b}{LCa}}, \quad
Q_p = RC \sqrt{\frac{1}{LC}}
$$

---

### Resultado final

$$
\boxed{
\begin{aligned}
k &= a \\\\
\omega_z &= \sqrt{\frac{b}{LCa}} \\\\
\omega_p &= \sqrt{\frac{1}{LC}} \\\\
Q_z &= \frac{RCa}{d} \sqrt{\frac{b}{LCa}} \\\\
Q_p &= RC \sqrt{\frac{1}{LC}}
\end{aligned}
}
$$

Con esto quedan establecidos los parámetros necesarios para implementar el circuito como una **sección biquadrática (SOS)**, cumpliendo con lo requerido en el enunciado.
